```python
%matplotlib notebook
import matplotlib.pyplot as pl
import numpy as np
from astropy.io import fits
```

```python
im = fits.open("./from_monika/D00145430_z_c01_r4148p01_immasked.fits.fz")
f = fits.open("./from_monika/D00145430_z_c01_r4148p01_psfex-starlist.fits")
```

```python
x = f[2].data['x_image'].astype(int)
y = f[2].data['y_image'].astype(int)
```

```python
goodstars= np.where(f[2].data['flags_psf']==0)[0]
```

```python
goodstars.shape
```

```python
box=10
N=100
count=0
fig, axs = pl.subplots(10,10,figsize=(50,50))
for coords in zip(x[goodstars][0:N],y[goodstars][0:N]):
    print(coords)
    cutout = im['sci'].data[ coords[1]-box:coords[1]+box , coords[0]-box:coords[0]+box ]
    axs.flatten()[count].imshow(cutout)
    axs.flatten()[count].set_xticks([])
    axs.flatten()[count].set_yticks([])
    count=count+1
pl.tight_layout()
pl.savefig('./test2.pdf')
```
